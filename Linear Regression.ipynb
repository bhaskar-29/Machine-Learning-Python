```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
%matplotlib inline
```

```python
df = pd.read_csv('canada_per_capita_income.csv')
df.head()
```

```python
plt.figure(figsize=(8,6))
plt.scatter(df['year'], df['per capita income (US$)'], color='red', s=20)
```

```python
import seaborn as sns
```

```python
sns.lmplot('year', 'per capita income (US$)', data=df)
```

```python
df_copy = df[df['year']<=2000]
```

```python
sns.lmplot('year', 'per capita income (US$)', data=df_copy)
```

```python
x = df.iloc[:,-2].values.reshape(-1,1)
y = df.iloc[:,-1].values.reshape(-1,1)
```

```python
reg = LinearRegression()
reg.fit(x, y)
```

```python
line = reg.coef_*x + reg.intercept_
plt.figure(figsize=(8,6))
plt.plot(x, line, color='blue')
plt.scatter(x, y, color='red')
```

```python
reg.predict([[2020]])
```
